Adding Labels to each of the datset


```python
df_benign['Label'] = "BENIGN"
df_bruteforce['Label'] = "BRUTEFORCE"
df_portscan['Label'] = "PORTSCAN"
```

Merging all datsets into one

```python
df =pd.concat([df_benign, df_bruteforce, df_portscan], ignore_index=True)
```

Removing duplicates and missing values


```python
df = df.dropna()
df = df.drop_duplicates()
```

Converting Labels to numeric values

```python
# ML algorithms work only with numeric data so we converted labels into numeric values

label = { "BENIGN":0, "BRUTEFORCE":1, "PORTSCAN":2}
df ['Label'] = df['Label'].map(label)
```
